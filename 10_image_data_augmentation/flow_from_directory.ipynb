```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array, ImageDataGenerator

gen = ImageDataGenerator(rotation_range=30, width_shift_range=0.3, shear_range=0.3, rescale=1./255)

data_path = '/content/drive/MyDrive/Git/tensorflow2-colab-ex/resources/img/test'
batch_siz = 3

# flow_from_directory() 함수는 이미지를 불러 올때 주어진 디렉토리의 하위 디렉토리 이름에 맞춰 자동으로 labelling 해줌
data_gen = gen.flow_from_directory(directory=data_path, 
                                   batch_size=batch_siz,
                                   shuffle=True, 
                                   target_size=(100, 100), 
                                   class_mode='categorical')
```

```python
img, label = data_gen.next() # batch_siz=3 지정된 개수만큼 이미지 생성

plt.figure(figsize=(6,6))

for i in range(len(img)):
  plt.subplot(1, len(img), i + 1)
  plt.xticks([])
  plt.yticks([])
  plt.title(str(np.argmax(label[i])))
  
  plt.imshow(img[i])
```
